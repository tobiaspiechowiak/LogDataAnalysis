```python
#define plot function for confusion matrix
def plot_confusion_matrix(y_test,y_pred,labels):
    cfm = confusion_matrix(y_test,y_pred) 
    #normalize
    cfm_norm = np.around(cfm / cfm.sum(axis=1),decimals = 2)
    #now plot stuff
    fig, ax = plt.subplots(figsize = (6, 6))
    ax.matshow(cfm_norm,cmap=plt.cm.Blues,alpha=0.7)

    ax.set(xticks=np.arange(cfm_norm.shape[1]),
    yticks=np.arange(cfm_norm.shape[0]),    
    xticklabels=labels, yticklabels=labels)
    plt.setp(ax.get_xticklabels(), rotation=45,rotation_mode = 'anchor', ha='left')
    
    ax.set_ylabel('True Class',fontsize=15)
    ax.set_xlabel('Predicted Class',fontsize=15)
    
    for i in range(0,cfm_norm.shape[0]):
        print(i)
        for j in range(0,cfm_norm.shape[1]):
            ax.text(x=j,y=i, s=cfm_norm[i,j],va='center', ha= 'center')
    return fig
```

```python
#clustering for Broadband / SNR 
from sklearn.mixture import GaussianMixture
import itertools
from scipy import linalg
import matplotlib as mpl
from skimage.measure import EllipseModel
from matplotlib.patches import Ellipse

names = ['broadband power','S_to_N'] 
input_left = data_left[names].values
input_right = data_right[names].values
X = np.hstack([input_left[:min(input_left.shape[0],input_right.shape[0]),:],
                    input_right[:min(input_left.shape[0],input_right.shape[0]),:]])
X_1,X_2 = X[:,:2], X[:,2:] #left, right
```

```python
#check the BIC and find the best cluster number (from scikit-learn.org)
lowest_bic = np.infty
bic = []
n_components_range = range(1, 10)
cv_type = 'full'
for n_components in n_components_range:
    # Fit a Gaussian mixture with EM
    gmm = GaussianMixture(n_components=n_components,
                                      covariance_type=cv_type)
    gmm.fit(X)
    bic.append(gmm.bic(X))
    if bic[-1] < lowest_bic:
        lowest_bic = bic[-1]
        best_gmm = gmm

min_index = np.argmin(bic)
```

```python
#make the clusters
clf = best_gmm
Y_ = clf.predict(X)
#Y2_ = clf.predict(X_2)

colors = ('blue','green','red','orange') 
face_colors = ('#D5E1F6','#A5D5A5','#F6D6D5','#F3CD8E') 
groups = ('Cluster 1','Cluster 2','Cluster 3','Cluster 4')

fig = plt.figure(figsize=(15,8))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#generate ellipse
ell = EllipseModel()

for i, (color, facecolor, group, cov, mean) in enumerate(zip(colors, face_colors, groups,clf.covariances_,clf.means_)):     
    
    #left side
    points = np.array((X[Y_ == i, 1], X[Y_ == i, 0]))
    ell.estimate(points.transpose())
    xc, yc, a, b, theta = ell.params
    ell_patch = Ellipse((xc, yc), 2*a, 2*b, theta*180/np.pi, edgecolor=color, facecolor=color,
                       alpha = 0.25)
    ax1.add_patch(ell_patch)
    #Points
    ax1.scatter(X[Y_ == i, 1], X[Y_ == i, 0], color=color,alpha=0.8,s=25,label = group,
                edgecolor = 'k')
    
    
    #right side
    points = np.array((X[Y_ == i, 3], X[Y_ == i, 2]))
    ell.estimate(points.transpose())
    xc, yc, a, b, theta = ell.params
    ell_patch = Ellipse((xc, yc), 2*a, 2*b, theta*180/np.pi, edgecolor=color, facecolor=color,
                        alpha = 0.25)
    ax2.add_patch(ell_patch)
    
    #Points  
    ax2.scatter(X[Y_ == i, 3], X[Y_ == i, 2], color=color,alpha=0.8,s=25,label = group,
                edgecolor = 'k')
     
     
    
ax1.axvline(x=0,linestyle='--')
#ax1.axvline(x=-10,linestyle='--')
ax1.set_xlabel('SNR (dB)',fontsize=18)
ax1.set_ylabel('Broadband level (dB SPL)',fontsize=18)
ax1.set_title('Left ear',fontsize=25)
ax1.set_xlim((-30,30))
ax1.set_ylim((30,90))
ax1.tick_params(axis='x',labelsize=15)
ax1.tick_params(axis='y',labelsize=15)

ax2.axvline(x=0,linestyle='--')
ax2.set_xlabel('SNR (dB)',fontsize=18)
ax2.set_ylabel('Broadband level (dB SPL)',fontsize=18)
ax2.set_title('Right ear',fontsize=25)
ax2.set_xlim((-30,30))
ax2.set_ylim((30,90))
ax2.tick_params(axis='x',labelsize=15)
ax2.tick_params(axis='y',labelsize=15)

ax1.legend(loc=3)
ax2.legend(loc=3)


```

```python
#BASS detector states 
bass_states_left = {}
bass_states_right = {}
names = ['as_mix_bassZeroOrOne_1','as_zoom_amount_1','class']
data_left_ = data_left[names]
data_right_ = data_right[names]

for idx in range(0,6):
    bass_states_left[idx] = data_left_[data_left_['class'] ==idx+1]
    bass_states_right[idx] = data_right_[data_right['class']==idx+1]
    
```

```python
def bass_outputs(as_mix,as_zoom):
    BES_monitor = (1 - as_zoom) * (1 - as_mix)
    FD = as_zoom
    MIE = (1 - as_zoom) * as_mix
    return (BES_monitor, FD, MIE)
        
```

```python
#compute bass states for each scene
x = np.arange(979)

a_left = {}
b_left = {}
c_left = {}
a_right = {}
b_right = {}
c_right = {}

for idx in range(0,6):
    (a_left[idx],b_left[idx],c_left[idx]) = bass_outputs(bass_states_left[idx].values[:,0],
                                        bass_states_left[idx].values[:,1])
    (a_right[idx],b_right[idx],c_right[idx]) = bass_outputs(bass_states_right[idx].values[:,0],
                                         bass_states_right[idx].values[:,1])
    
```
